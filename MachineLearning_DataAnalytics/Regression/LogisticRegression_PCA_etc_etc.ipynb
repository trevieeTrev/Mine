**Answer**:


The Bayes Naive classifier selects the most likely classification Vnb given the attribute values $a_1 , a_2 , . . . a_n$.
This results in:

$$V_{n b}=\operatorname{argmax}_{v_{j} \in V} P\left(v_{j}\right) \prod P\left(a_{i} | v_{j}\right)$$

We generally estimate $P(a_i|v_j)$ using m-estimates:

$$P\left(a_{i} | v_{j}\right)=\frac{n_{c}+m p}{n+m}$$

\begin{array}{l}{n=\text { the number of training examples for which } v=v_{j}} \\ {n_{c}=\text { number of examples for which } v=v_{j} \text { and } a=a_{i}} \\ {p=\text { a prior estimate for } P\left(a_{i} | v_{j}\right)} \\ {m=\text { the equivalent sample size }}\end{array}

We can estimate these values using the above equation.

Take one value $P (Red|Yes)$ for example. We have 5 cases where $v_j = Yes$ , and in 2 of those cases $a_i = Red$. So for $P (Red|Y es)$, $n = 5$ and $n_c = 2$. Note that all attribute are binary (two possible values). We are assuming no other information so,$ p = 1 / (number-of-attribute-values) = 0.5$ for all of our attributes. Our $m$ value is arbitrary, (We will use $m = 3$) but consistent for all attributes.

For the samples with label is "YES":
\begin{array} 
{l}Red: SUV: Domestic: \\{n=5}  \quad {n=5}  \quad {n=5}\\ {n_c=2}  \quad {n_c=1}  \quad {n_c=2} \\ {p=0.5}  \quad {p=0.5}  \quad {p=0.5}\\ {m=3} \quad {m=3} \quad {m=3}
\end{array}

For the samples with label is "NO":
\begin{array} 
{l}Red: SUV: Domestic: \\{n=5}  \quad {n=5}  \quad {n=5}\\ {n_c=2}  \quad {n_c=3}  \quad {n_c=3} \\ {p=0.5}  \quad {p=0.5}  \quad {p=0.5}\\ {m=3} \quad {m=3} \quad {m=3}
\end{array}

Based on the above information, the probabilities will be: 
$$P(\operatorname{Red} | Y e s)=\frac{2+3 * 0.5}{5+3}=.44$$
$$
P(S U V | Y e s)=\frac{1+3 * 0.5}{5+3}=.31

$$

We have  $P(Y e s)=0.5 \text { and } P(No)=0.5$, so we can apply the above equation.

For  $P(Yes)=0.5$,

### Image compression
The greater the eigenvalues -> the great the variance -> most information for the eigenvector.

SVD each matrix and get rid of the small eigenvlues, the image is compressed. The loss of inforamtion is negligible as the difference is very difficult to be spotted. 

[Principal Component Analysis](https://en.wikipedia.org/wiki/Principal_component_analysis) follows the same process to elimate the small variance eigenvalues and their vectors.

```python
def buildTrainX(images):
    number_dataset = 32
    trainX = np.zeros((number_dataset,64*64))
    for i in range(number_dataset):  
        gray = rgb2gray(images[i])
        tempX = np.reshape(gray, [64*64])
        trainX[i,:] = tempX
    return trainX

trainX = buildTrainX(images)
```

You need to finish the following functions to do SVD and then reconstruct the image by components.

```python
def centerX(trainX):
    """
    Compute the mean for each coloum at trainX, and then for each item in trainX, minus the correspond mean.
    """
    Xcenters = np.mean(trainX,axis=0)
    X = trainX - Xcenters
#     X = trainX
    return X

def svd(A):
    """
    Do SVD. You could use numpy SVD.
    """
    U, S, V = np.linalg.svd(A)
    return U, S, V

def rebuildsvd(U, S, V, k):
    """
    Rebuild SVD by k componments.
    """
#   --method1: 
#     r = len(S)
#     Xrebuild = np.dot(U[:,:r] * S, V[:r,:]) 

#   --method2:
    u1 = U[:,0:k]
    s1 = S[0:k]
    s1 = np.diag(s1)
    t1 = np.dot(u1,s1)
    v1 = V[0:k,:]
    Xrebuild = np.dot(t1,v1)

    return Xrebuild
```

Now let see the results for image reconstruction.

```python
X = centerX(trainX)
U, S, V = svd(X)
component_num = [2,5,10,20,32]

fig = plt.figure(figsize=(16, 16))
# plot several images
i=0
for k in component_num:
    Xrebuild = rebuildsvd(U, S, V, k)
    img_rebuild = np.reshape(Xrebuild[0,:] + np.mean(trainX,axis=0), [64,64])
    ax = fig.add_subplot(5, 5, i + 1, xticks=[], yticks=[])
    ax.imshow(img_rebuild, cmap=plt.cm.bone)
    i = i+1
```

```python
def rmse(pred, label): # This is the root mean square error (use the firs equation here)
    sha = np.shape(pred)
    length = sha[0];
    out = np.sqrt((1/length)*np.sum(np.square(pred-label)))
    return out
#     return np.sqrt(np.mean((pred - label) ** 2))



# Hint: get the sqaure root of theta hat in slide 14 (https://mahdi-roozbahani.github.io/cse4240-spring2019.github.io/course/14-linear-regression.pdf)
prediction = np.array([1, 2, 3]) 
label = np.array([1.5, 2.5, 3.5])
print('rmse =',rmse(prediction, label))
```

```python
def construct_polynomial_feats(x, degree):
    """
    Args:
        x: numpy array of length N, the 1-D observations
        degree: the max polynomial degree
    Return:
        feat: numpy array of shape Nx(degree+1), remember to include 
        the bias term. feat is in the format of:
        [[1.0, x1, x1^2, x1^3, ....,],
         [1.0, x2, x2^2, x2^3, ....,],
         ......
        ]
    """
    feat_list = [np.ones(x.shape), x]
    cur_x = x.copy()
    for d in range(2, degree + 1):
        cur_x = cur_x * x
        feat_list.append(cur_x)

    return np.stack(feat_list, axis=1)

# Here there are two data points one data point is 0.5 and other one is 0.2. We try to represent these two data points
# in higher dimensions
x = np.array([0.5, 0.2]) 
x_feat = construct_polynomial_feats(x, 4)
print(x_feat)
```

```python
class LinearReg(object):
    @staticmethod 
    # static method means that you can use this method or function for any other classes, it is not specific to LinearReg
    def fit(xtrain, ytrain):
        """
        Args:
            xtrain: NxD numpy array, where N is number 
                    of instances and D is the dimensionality of each 
                    instance
            ytarin: Nx1 numpy array, the true labels
        Return:
            weight: Dx1 numpy array, the weights of linear regression model
        """
#         t = np.dot(xtrain.T, xtrain)
#         t = np.linalg.inv(t)
        
#         c = np.dot(xtrain.T, ytrain)
#         weight = np.dot(t, c)
#         return weight
    
        xtx = np.matmul(np.transpose(xtrain),xtrain)
        inv = np.linalg.inv(xtx)
        pseudoinv = np.matmul(inv,np.transpose(xtrain))
        weight = np.matmul(pseudoinv,ytrain)
        return weight

    @staticmethod
    def predict(xtest, weight):
        """
        Args:
            xtest: NxD numpy array, where N is number 
                   of instances and D is the dimensionality of each 
                   instance
            weight: Dx1 numpy array, the weights of linear regression model
        Return:
            prediction: Nx1 numpy array, the predicted labels
        """
        prediction = np.dot(xtest, weight)
        return prediction
    
#         weight = np.transpose(weight)
#         prediction = weight*xtest #broadcasting along the rows
#         prediction = np.sum(prediction, axis = 1)
#         return prediction

# Hint: in the fit function, use close form solution of the linear regression to get weights. 
# For inverse, you can use numpy linear algebra function  
# For the predict, you need to use linear combination of data points and their weights (y = w0*1+w1*X1+...)
```

Now let's try Ridge Regression

```python
class RidgeReg(LinearReg):

    @staticmethod
    def fit(xtrain, ytrain, c_lambda):
        dim = xtrain.shape[1]
#         print(dim)
        t1 = np.dot(xtrain.T, xtrain)
        I = np.eye(dim)
        I[0, 0] = 0
        t2 = c_lambda * I + t1
        t2inv = np.linalg.inv(t2)
        t3 = np.dot(t2inv, xtrain.T)
        beta = np.dot(t3, ytrain)
        return beta
#        
```

```python
# Let's use Cross Validation to find what can be the best value for the c_lambda. 
# We provided 6 possible values for lambda, and you will use them in cross validation.
# For cross validation, use 10-fold method and only use it for your training data (you already have the train_indices to get training data).
# For the tarining data, split them in 10 folds which means that use 10 percent of training data for test and 90 percent for training.
# At the end for each lambda, you have caluclated 10 rmse and get the mean value of that.
# That's it. Pick up the lambda with the lowest mean value of rmse. 
def cross_validation(X, y, kfold, c_lambda):
    num_per_fold = X.shape[0] // kfold
    errors = []
    for i in range(kfold):
        valid_x = X[i * num_per_fold : (i + 1) * num_per_fold]
        valid_y = y[i * num_per_fold : (i + 1) * num_per_fold]

        x1 = X[:i * num_per_fold]
        x2 = X[(i + 1) * num_per_fold:]
        train_x = np.concatenate([x1, x2], axis=0)
        
        y1 = y[:i * num_per_fold]
        y2 = y[(i + 1) * num_per_fold:]
        train_y = np.concatenate([y1, y2], axis=0)
        weight = RidgeReg.fit(train_x, train_y, c_lambda=c_lambda)
        valid_pred = RidgeReg.predict(valid_x, weight) 
        error = rmse(valid_pred, valid_y)
        errors.append(error)
        meanErrors = np.mean(errors)
    return meanErrors


best_lambda = None
best_error = None
kfold = 10
lambda_list = [0, 0.1, 1, 5, 10, 100, 1000]
for lm in lambda_list:
    err = cross_validation(x_all_feat[train_indices], y_all[train_indices], kfold, lm)
    print('lambda: %.2f' % lm, 'error: %.6f'% err)
    if best_error is None or err < best_error:
        best_error = err
        best_lambda = lm

print('best_lambda: %.2f' % best_lambda)
weight = RidgeReg.fit(x_all_feat[train_indices], y_all[train_indices], c_lambda=best_lambda)
y_test_pred = RidgeReg.predict(x_all_feat[test_indices], weight)
test_rmse = rmse(y_test_pred, y_all[test_indices])
print('test rmse: %.4f' % test_rmse)
```
