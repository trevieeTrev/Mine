```python
# 2(A){i}

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-.5, 1.5, 30)
y = np.exp(x)
# x = np.array(range(-2,2))
# y =  np.exp(x)

x1 = np.linspace(-.5, 1.5, 30)
y1 = x + 1


plt.plot(x,y,"b-", label='e$^x$')
plt.plot(x1,y1,"r-", label='1+x')


plt.legend(bbox_to_anchor=(1.5, 1), loc=0, borderaxespad=0)

plt.show()
```

```python
#Question 2(A) {ii}
import matplotlib.pyplot as plt
import numpy as np


x = np.array(range(10))
y =  x

x1 = np.array(range(10))
y1 = np.log(1+x)


plt.plot(x,y,"b-", label='x')
plt.plot(x1,y1,"r-", label='log(1+x)')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()


```

```python
#Question 2(A) {iii}

import matplotlib.pyplot as plt
import numpy as np

for a in range(1,2):
    xl = [i for i in np.arange(-1,2,.01)]
    x = np.asarray(xl)
    y = 1 + (np.exp(a) - 1)*x

    x1l = [i for i in np.arange(-1,2,0.08)]
    x1 = np.asarray(x1l)
    y1 = np.exp(a*x1)


    plt.plot(x,y,"b-", label='e^ax')
    plt.plot(x1,y1,"r-", label='1+(e$^a$ - 1)x')


    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.xlabel('a = {}'.format(a))
    plt.show()
```

```python
#Question 2(A) {iv}
import matplotlib.pyplot as plt
import numpy as np


xl = [i for i in np.arange(0,0.5,0.05)]
x = np.asarray(xl)
y = x + (x**2)

x1l = [i for i in np.arange(0,0.5,0.05)]
x1 = np.asarray(x1l)
y1 = np.log(1-x)*-1


plt.plot(x,y,"b-", label='x + x$^2$')
plt.plot(x1,y1,"r-", label='-log(1-x)')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
